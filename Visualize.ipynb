```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_userdata = pd.read_csv("USER_DATA.csv", encoding="utf-8-sig")
df_booking = pd.read_csv("BOOKING.csv", encoding="utf-8-sig")
df_clubdata = pd.read_csv("CLUB_DATA.csv", encoding="utf-8-sig")
df_courtdata = pd.read_csv("COURT_DATA.csv", encoding="utf-8-sig")
df_operation= pd.read_csv("OPERATION_DATA.csv", encoding="utf-8-sig")
```

```python
sns.set(style="whitegrid", font_scale=1.1)
plt.rcParams["figure.figsize"] = (10, 6)
```

# USER_DATA

```python
# AGE DISTRIBUTION
plt.figure()
sns.histplot(df_userdata["age"], bins=20, kde=True, color="skyblue")
plt.title("Phân bố độ tuổi người dùng")
plt.xlabel("Tuổi")
plt.ylabel("Số lượng")
plt.show()
```

```python
# visualize gender distribution
plt.figure()
sns.countplot(x="gender", data=df_userdata, palette="pastel")   
plt.title("Phân bố giới tính người dùng")
plt.xlabel("Giới tính")
plt.ylabel("Số lượng")
plt.show()
```

```python
# visualize user district distribution
plt.figure()
sns.countplot(y="user_district", data=df_userdata, order=df_userdata["user_district"].value_counts().index, palette="muted")
plt.title("Phân bố quận người dùng")
plt.xlabel("Số lượng")
plt.ylabel("Quận")
plt.show()
```
